```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

DATASET: Employment rate of adults by sex, age groups, educational attainment level, number of children and age of youngest child

```python
df_em = pd.read_csv("Employment rate of adults by sex, age groups, educational attainment level, number of children and age of youngest child (%) (lfst_hheredch).tsv", sep='/t')
df_em.head()
```

```python
df_em_sep = df_em.iloc[:, 0].str.split(',', expand=True)
df_em_sep.columns = ['freq', 'sex', 'age', 'n_child', 'agechild', 'isced11', 'unit', 'geo_years']  # 'geo_years' contiene los años separados por tabs
df_em_sep.head()
```

```python
df_em_sep["age"].unique()
```

```python
df_em_sep["sex"].unique()
```

```python
# Paso 1: Separar la columna 'geo_years' en dos partes: 'geo' y los datos de años
df_em_sep[['geo', 'years_data']] = df_em_sep ['geo_years'].str.split('\t', n=1, expand=True)

# Eliminar la columna 'geo_years' ya que ahora está dividida
df_em_sep.drop(columns=['geo_years'], inplace=True)

# Paso 2: Separar la columna 'years_data' en columnas individuales para cada año
years_split = df_em_sep['years_data'].str.split('\t', expand=True)

# Paso 3: Asignar nombres a las columnas de años, desde 2006 hasta 2023
years_split.columns = [str(year) for year in range(2006, 2023 + 1)]

# Paso 4: Eliminar la columna 'years_data' ya que no la necesitamos más
df_em_sep.drop(columns=['years_data'], inplace=True)

# Paso 5: Concatenar las columnas iniciales con las nuevas columnas de años
df_em_sep_final = pd.concat([df_em_sep[['freq', 'sex', 'age', 'n_child', 'agechild', 'isced11', 'unit', 'geo']], years_split], axis=1)

# Verificar el resultado final
df_em_sep_final.head()
```

```python
# Mantener sólo las columnas de los años 2014 a 2023 junto con las columnas iniciales
df_em_sep_final = df_em_sep_final[['freq', 'sex', 'age', 'n_child', 'agechild', 'isced11', 'unit', 'geo'] + [str(year) for year in range(2014, 2023 + 1)]]

# Verificar el resultado
df_em_sep_final.head()
```

```python

df_employ = df_em_sep_final.sort_values(by='geo').reset_index(drop=True)

# Paso 2: Reestructurar el dataset (unpivoting)
df_final = df_employ.melt(
    id_vars=['freq', 'sex', 'age', 'n_child', 'agechild', 'isced11', 'unit', 'geo'],  # Mantener estas columnas
    value_vars=[str(year) for year in range(2014, 2024)],  # Años como columnas
    var_name='year',  # Nueva columna para los años
    value_name='rate of adult employment (%)'  # Nueva columna para los valores
)

# Mostrar el resultado
df_final.head()
```

EMPEZAMOS A LIMPIAR LAS COLUMNAS:

```python

# Reemplazar las categorías por valores numéricos
df_final['agechild'] = df_final['agechild'].replace({
    'NAP': "sin valor",  
    'Y_LT6': "< 6",  
    'Y6-11': "6-11", 
    'TOTAL': "todas las edades",  
    'Y_GE12': ">= 12"
})

df_final ["agechild"].unique()
```

```python
df_final["isced11"] = df_final["isced11"].replace({
    'ED0-2': "bajo", 
    'ED5-8': "alto", 
    'ED3_4': "medio", 
    'TOTAL': "todos los niveles", 
    'UNK': "desconocido", 
    'NRP': "muestra irrelevante"})
df_final = df_final.rename(columns={"isced11": "nivel_educativo"})
df_final["nivel_educativo"].unique()
```

```python
df_final["n_child"] = df_final["n_child"].replace({'0': 0, 'GE3': " 3 o más", 'TOTAL': "todos", '1': 1, '2': 2})
df_final["n_child"].unique()
```

```python
df_final.head(10)
```

```python
# Separar los valores numéricos y las banderas
df_final[['rate_value', 'rate_flag']] = df_final['rate of adult employment (%)'].str.extract(r'([\d.]+)|([a-zA-Z]+)')

# Llenar la columna 'rate_flag' en caso de que esté vacía y la celda original solo tenga bandera
df_final['rate_flag'] = df_final['rate_flag'].fillna(
    df_final['rate of adult employment (%)'].str.extract(r'([a-zA-Z]+)')[0]
)

# Convertir la columna 'rate_value' a tipo float
df_final['rate_value'] = pd.to_numeric(df_final['rate_value'], errors='coerce')

```

CREAR NUEVAS COLUMNAS PARA EL ANÁLISIS

```python
# Asegurarse de que la columna 'year' sea numérica
df_final['year'] = pd.to_numeric(df_final['year'], errors='coerce')

# Ahora puedes realizar la operación de años desde 2014
df_final['years_since_2014'] = df_final['year'] - 2014

df_final.head(10)
```

```python
df_final["years_since_2014"].unique()
```

```python
df_final['employment_rate_category'] = pd.cut(df_final['rate_value'], bins=[0, 50, 75, 100], labels=["bajo", "medio", "alto"])
df_final.head(10)
```

```python
# Reemplazar valores en la columna 'age' por tuplas (min, max)
df_final["age"] = df_final["age"].replace({
    'Y18-24': "18, 24",
    'Y25-49': "25, 49",
    'Y_GE55': "55, None",
    'Y18-64': "18, 64",
    'Y25-54': "25, 54",
    'Y_GE18': "18, None",
    'Y20-49': "20, 49"
})

# Verificar los cambios
print(df_final["age"].unique())
```

2. Imputación de valores NaN en la columna "age:max":
voy a imputar los datos de age_max igualándolos a la edad de jubilación para hombres y mujeres en los respectivos países de las columnas


```python
# Filtrar filas con NaN en la columna 'age_max'
nan_rows = df_final[df_final['age_max'].isna()]
print(nan_rows[['geo', 'sex', 'age_max']])
```

* creo dos diccionarios con los valores encontrados en bibliografía para las edades de jubilación por país tanto en hombres como en mujeres:

```python
# Diccionario de edades de jubilación para hombres
retirement_age_men = {
    'AT': 65, 'BA': 65, 'BE': 65, 'BG': 64.58, 'CY': 65.5, 'CZ': 64.33, 
    'DE': 66, 'DK': 67, 'EE': 64.75, 'EL': 67, 'ES': 66.5, 'FI': 64.5, 
    'FR': 62.5, 'HR': 65, 'HU': 65, 'IE': 66, 'IT': 67, 'LT': 64.67, 
    'LU': 65, 'LV': 64.75, 'ME': 66.67, 'MK': 64, 'MT': 64, 'NL': 67, 
    'NO': 62, 'PL': 65, 'PT': 66.33, 'RO': 65, 'RS': 65, 'SE': 63, 
    'SI': 60, 'SK': 63, 'TR': 60, 'UK': 66
}

# Diccionario de edades de jubilación para mujeres
retirement_age_women = {
    'AT': 60.5, 'BA': 65, 'BE': 65, 'BG': 62.17, 'CY': 65.5, 'CZ': 64.33, 
    'DE': 66, 'DK': 67, 'EE': 64.75, 'EL': 67, 'ES': 66.5, 'FI': 64.5, 
    'FR': 62.5, 'HR': 63.5, 'HU': 65, 'IE': 66, 'IT': 67, 'LT': 64.33, 
    'LU': 65, 'LV': 64.75, 'ME': 62.5, 'MK': 64, 'MT': 64, 'NL': 67, 
    'NO': 62, 'PL': 60, 'PT': 66.33, 'RO': 62, 'RS': 63.67, 'SE': 66, 
    'SI': 60, 'SK': 63, 'TR': 58, 'UK': 66
}
```

```python
# Función para rellenar NaN en 'age_max' según 'geo' y 'sex'
def fill_age_max(row):
    if row['sex'] == 'M':  # Si es hombre
        return retirement_age_men.get(row['geo'], row['age_max'])  # Rellenar con el diccionario de hombres
    elif row['sex'] == 'F':  # Si es mujer
        return retirement_age_women.get(row['geo'], row['age_max'])  # Rellenar con el diccionario de mujeres
    return row['age_max']  # Si no es ni 'M' ni 'F', no hacer nada

# Rellenar los NaN de la columna 'age_max' con los valores de los diccionarios
df_final['age_max'] = df_final.apply(lambda row: fill_age_max(row) if pd.isna(row['age_max']) else row['age_max'], axis=1)
```

```python
# Verificar si todavía quedan NaN en 'age_max'
df_final[df_final['age_max'].isna()].head(10)
```

```python
# creamos una función para categorizar la columna 'agechild' y así tener los datos más trabajables:
def categorize_age_group(age):
    if isinstance(age, str):
        if 'sin valor' in age:
            return 0
        elif '6-11' in age:
            return 'niños'
        elif '>= 12' in age:
            return 'adolescentes'
        elif '< 6' in age:
            return 'infantes'
    return age  # Retorna el valor original si no hay coincidencia

# Aplica la categorización a la columna 'agechild'
df_final['agechild'] = df_final['agechild'].apply(categorize_age_group)
```

AHORA VAMOS A IMPUTAR LOS DATOS FALTANTES EN LA COLUMNA RATE_VALUE CON LA MEDIANA DE LA COLUMNA AGRUPADA POR PAÍS Y POR AÑO

```python
df_final.isna().sum()
```

```python
# Primero, calcula las medianas para los grupos válidos
medianas = df_final.groupby(['geo', 'year'])['rate_value'].median()
```

```python
# Rellenar los NaN con las medianas correspondientes, manejando posibles claves faltantes
def fill_rate_value(row):
    if pd.isna(row['rate_value']):
        key = (row['geo'], row['year'])
        # Verifica si la clave existe en el diccionario de medianas
        return medianas[key] if key in medianas else row['rate_value']
    return row['rate_value']

df_final['rate_value'] = df_final.apply(fill_rate_value, axis=1)
```

```python
df_final.isna().sum()
```

A PESAR DE LA IMPUTACIÓN DE LOS VALORES NaN CON LAS MEDIANAS CORRESPONDIENTES POR PAÍS Y AÑO, VEMOS QUE AÚN TENEMOS VALORES NaN QUE NO HAN PODIDO ADECUARSE A LA FUNCIÓN CREADA POR QUE SIGUEN FALTANDO DATOS, ASÍ QUE USAREMOS OTRA ESTRATEGIA DE IMPUTACIÓN DE VALORES:

```python
# IDENTIFICAMOS LAS MEDIANAS FALTANTES DE ACUERDO AL PAÍS Y AÑO
missing_medians = df_final[pd.isna(df_final['rate_value'])][['geo', 'year']].drop_duplicates()
print(missing_medians)
```

OBSERVAMOS QUE TENEMOS DATOS FALTANTES DE CIERTOS PAÍSES, tienen un patrón claro: pertenecen a ciertos países y años donde aparentemente no se tienen datos válidos para calcular la mediana de rate_value

En lugar de una mediana global, VAMOS A calcular una mediana específica por país y usarla para rellenar las combinaciones faltantes. Esto captura mejor las diferencias entre países:

```python
median_by_country = df_final.groupby('geo')['rate_value'].median()
df_final['rate_value'] = df_final.apply(
    lambda row: median_by_country[row['geo']] if pd.isna(row['rate_value']) else row['rate_value'],
    axis=1
)
```
