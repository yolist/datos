```python
import pandas as pd
```

DATASET: Employment rate of adults by sex, age groups, educational attainment level, number of children and age of youngest child

```python
df_em = pd.read_csv("Employment rate of adults by sex, age groups, educational attainment level, number of children and age of youngest child (%) (lfst_hheredch).tsv", sep='/t')
df_em.head()
```

```python
df_em_sep = df_em.iloc[:, 0].str.split(',', expand=True)
df_em_sep.columns = ['freq', 'sex', 'age', 'n_child', 'agechild', 'isced11', 'unit', 'geo_years']  # 'geo_years' contiene los a√±os separados por tabs
df_em_sep.head()
```

```python
df_em_sep["age"].unique()
```

```python
df_em_sep["sex"].unique()
```

```python
# Paso 1: Separar la columna 'geo_years' en dos partes: 'geo' y los datos de a√±os
df_em_sep[['geo', 'years_data']] = df_em_sep ['geo_years'].str.split('\t', n=1, expand=True)

# Eliminar la columna 'geo_years' ya que ahora est√° dividida
df_em_sep.drop(columns=['geo_years'], inplace=True)

# Paso 2: Separar la columna 'years_data' en columnas individuales para cada a√±o
years_split = df_em_sep['years_data'].str.split('\t', expand=True)

# Paso 3: Asignar nombres a las columnas de a√±os, desde 2006 hasta 2023
years_split.columns = [str(year) for year in range(2006, 2023 + 1)]

# Paso 4: Eliminar la columna 'years_data' ya que no la necesitamos m√°s
df_em_sep.drop(columns=['years_data'], inplace=True)

# Paso 5: Concatenar las columnas iniciales con las nuevas columnas de a√±os
df_em_sep_final = pd.concat([df_em_sep[['freq', 'sex', 'age', 'n_child', 'agechild', 'isced11', 'unit', 'geo']], years_split], axis=1)

# Verificar el resultado final
df_em_sep_final.head()
```

```python
# Mantener s√≥lo las columnas de los a√±os 2014 a 2023 junto con las columnas iniciales
df_em_sep_final = df_em_sep_final[['freq', 'sex', 'age', 'n_child', 'agechild', 'isced11', 'unit', 'geo'] + [str(year) for year in range(2014, 2023 + 1)]]

# Verificar el resultado
df_em_sep_final.head()
```

```python

df_employ = df_em_sep_final.sort_values(by='geo').reset_index(drop=True)

# Paso 2: Reestructurar el dataset (unpivoting)
df_final = df_employ.melt(
    id_vars=['freq', 'sex', 'age', 'n_child', 'agechild', 'isced11', 'unit', 'geo'],  # Mantener estas columnas
    value_vars=[str(year) for year in range(2014, 2024)],  # A√±os como columnas
    var_name='year',  # Nueva columna para los a√±os
    value_name='rate of adult employment (%)'  # Nueva columna para los valores
)

# Mostrar el resultado
df_final.head()
```

EMPEZAMOS A LIMPIAR LAS COLUMNAS:

```python

# Reemplazar las categor√≠as por valores num√©ricos
df_final['agechild'] = df_final['agechild'].replace({
    'NAP': "sin valor",  
    'Y_LT6': "< 6",  
    'Y6-11': "6-11", 
    'TOTAL': "todas las edades",  
    'Y_GE12': ">= 12"
})

df_final ["agechild"].unique()
```

```python
df_final["isced11"] = df_final["isced11"].replace({
    'ED0-2': "bajo", 
    'ED5-8': "alto", 
    'ED3_4': "medio", 
    'TOTAL': "todos los niveles", 
    'UNK': "desconocido", 
    'NRP': "muestra irrelevante"})
df_final = df_final.rename(columns={"isced11": "nivel_educativo"})
df_final["nivel_educativo"].unique()
```

```python
df_final["n_child"] = df_final["n_child"].replace({'0': 0, 'GE3': " >=3", 'TOTAL': "todos", '1': 1, '2': 2})
df_final["n_child"].unique()
```

```python
df_final.head(10)
```

```python
# Separar los valores num√©ricos y las banderas
df_final[['rate_value', 'rate_flag']] = df_final['rate of adult employment (%)'].str.extract(r'([\d.]+)|([a-zA-Z]+)')

# Llenar la columna 'rate_flag' en caso de que est√© vac√≠a y la celda original solo tenga bandera
df_final['rate_flag'] = df_final['rate_flag'].fillna(
    df_final['rate of adult employment (%)'].str.extract(r'([a-zA-Z]+)')[0]
)

# Convertir la columna 'rate_value' a tipo float
df_final['rate_value'] = pd.to_numeric(df_final['rate_value'], errors='coerce')

```

```python
df_final.head(10)
```

CREAR NUEVAS COLUMNAS PARA EL AN√ÅLISIS

```python
# Asegurarse de que la columna 'year' sea num√©rica
df_final['year'] = pd.to_numeric(df_final['year'], errors='coerce')

# Ahora puedes realizar la operaci√≥n de a√±os desde 2014
df_final['years_since_2014'] = df_final['year'] - 2014

df_final.head(10)
```

```python
df_final["years_since_2014"].unique()
```

```python
df_final['employment_rate_category'] = pd.cut(df_final['rate_value'], bins=[0, 50, 75, 100], labels=["bajo", "medio", "alto"])
df_final.head(10)
```

```python
# Reemplazar valores en la columna 'age' por tuplas (min, max)
df_final["age"] = df_final["age"].replace({
    'Y18-24': "18, 24",
    'Y25-49': "25, 49",
    'Y_GE55': "55, None",
    'Y18-64': "18, 64",
    'Y25-54': "25, 54",
    'Y_GE18': "18, None",
    'Y20-49': "20, 49"
})

# Verificar los cambios
print(df_final["age"].unique())
```

```python
# Separar las cadenas de texto en dos columnas 'age_min' y 'age_max'
df_final[['age_min', 'age_max']] = df_final['age'].str.split(', ', expand=True)

# Convertir las columnas 'age_min' y 'age_max' a tipo num√©rico, para manejar valores como 'None'
df_final['age_min'] = pd.to_numeric(df_final['age_min'], errors='coerce')
df_final['age_max'] = pd.to_numeric(df_final['age_max'], errors='coerce')

df_final.head(10)
```

REORDENAR COLUMNAS Y ELIMINAR REPETIDAS:

```python
# Reordenar las columnas
new_order = ['age', 'rate of adult employment (%)', 'geo', 'sex','freq', 'age_min', 'age_max', 'year', 'n_child', 'agechild', 'nivel_educativo', 'unit', 'rate_value', 'rate_flag', 'years_since_2014', 'employment_rate_category']
df_final = df_final[new_order]

# Eliminar las columnas innecesarias (por ejemplo 'age' y 'rate of adult employment')
df_final = df_final.drop(columns=['age', 'rate of adult employment (%)'])

df_final.head(10)
```
