 #### 1ER DATASET: LONG TERM UNEMPLOYMENT RATE BY SEX

```python
import pandas as pd
```

```python
df_ltunemployment = pd.read_csv("C:/Users/YOLANDA/Desktop/TFM/personal/Long-term unemployment rate by sex.tsv", sep='/t')
df_ltunemployment.head() 
```

```python
# Paso 0: Separar 'freq', 'unit', 'age', 'sex', 'geo' con comas en primer lugar
df_ltunemployment_separated = df_ltunemployment.iloc[:, 0].str.split(',', expand=True)
df_ltunemployment_separated.columns = ['freq', 'indic_em', 'age', 'sex', 'unit', 'geo_years']  # 'geo_years' contiene los años separados por tabs
df_ltunemployment_separated.head()
```

```python
# Paso 1: Separar la columna 'geo_years' en dos partes: 'geo' y los datos de años
df_ltunemployment_separated[['geo', 'years_data']] = df_ltunemployment_separated['geo_years'].str.split('\t', n=1, expand=True)

# Eliminar la columna 'geo_years' ya que ahora está dividida
df_ltunemployment_separated.drop(columns=['geo_years'], inplace=True)

# Paso 2: Separar la columna 'years_data' en columnas individuales para cada año
years_split = df_ltunemployment_separated['years_data'].str.split('\t', expand=True)

# Paso 3: Asignar nombres a las columnas de años, desde 2005 hasta 2023
years_split.columns = [str(year) for year in range(2005, 2023 + 1)]

# Paso 4: Eliminar la columna 'years_data' ya que no la necesitamos más
df_ltunemployment_separated.drop(columns=['years_data'], inplace=True)

# Paso 5: Concatenar las columnas iniciales con las nuevas columnas de años
df_ltunemployment_final = pd.concat([df_ltunemployment_separated[['freq', 'unit', 'age', 'sex', 'geo']], years_split], axis=1)

# Verificar el resultado final
df_ltunemployment_final.head()
```

```python
df_ltunemployment_final["sex"].unique()
```

```python
# Mantener sólo las columnas de los años 2014 a 2023 junto con las columnas iniciales
df_ltunemployment_final = df_ltunemployment_final[['freq', 'age', 'unit', 'sex', 'geo'] + [str(year) for year in range(2014, 2023 + 1)]]

# Verificar el resultado
df_ltunemployment_final.head()
```

```python
# Obtener el orden actual de las columnas
cols = df_ltunemployment_final.columns.tolist()

# Intercambiar 'unit' y 'age'
cols[cols.index('unit')], cols[cols.index('age')] = cols[cols.index('age')], cols[cols.index('unit')]

# Reordenar el DataFrame con el nuevo orden de columnas
df_ltunemployment_final = df_ltunemployment_final[cols]

# Verificar el resultado
df_ltunemployment_final.head()
```

```python
df_ltun = df_ltunemployment_final.sort_values(by='geo').reset_index(drop=True)

# Paso 2: Reestructurar el dataset (unpivoting)
df_melted = df_ltun.melt(
    id_vars=['freq', 'unit','age', 'sex', 'geo'],  # Mantener estas columnas
    value_vars=[str(year) for year in range(2014, 2024)],  # Años como columnas
    var_name='year',  # Nueva columna para los años
    value_name='rate of long term unemployment (%)'  # Nueva columna para los valores
)

# Mostrar el resultado
df_melted.head()
```

```python
df_melted.shape
```

#### 3er dataset: Employment rates by sex, age, migration status, citizenship and educational attainment level

```python
df_erate = pd.read_csv("Employment rates by sex, age, migration status, citizenship and educational attainment level (lfsa_erganedm).tsv",  sep='/t')
df_erate.head()
```

```python
df_erate = df_erate.iloc[:, 0].str.split(',', expand=True)
df_erate.columns = ['freq', 'isced11', 'citizen', 'mgstatus', 'age', 'sex', 'unit', 'geo_years']  # 'geo_years' contiene los años separados por tabs
df_erate.head()
```

```python
# Paso 1: Separar la columna 'geo_years' en dos partes: 'geo' y los datos de años
df_erate[['geo', 'years_data']] = df_erate ['geo_years'].str.split('\t', n=1, expand=True)

# Eliminar la columna 'geo_years' ya que ahora está dividida
df_erate.drop(columns=['geo_years'], inplace=True)

# Paso 2: Separar la columna 'years_data' en columnas individuales para cada año
years_split = df_erate['years_data'].str.split('\t', expand=True)

# Paso 3: Asignar nombres a las columnas de años, desde 2021 hasta 2023
years_split.columns = [str(year) for year in range(2021, 2023 + 1)]

# Paso 4: Eliminar la columna 'years_data' ya que no la necesitamos más
df_erate.drop(columns=['years_data'], inplace=True)

# Paso 5: Concatenar las columnas iniciales con las nuevas columnas de años
df_erate_final = pd.concat([df_erate[['freq', 'isced11', 'citizen', 'mgstatus', 'age', 'sex', 'unit', 'geo']], years_split], axis=1)

# Verificar el resultado final
df_erate_final.head()
```

```python
df_emplrate = df_erate_final.sort_values(by='geo').reset_index(drop=True)

# Paso 2: Reestructurar el dataset (unpivoting)
df_emplrate.melt(
    id_vars=['freq', 'isced11', 'citizen', 'mgstatus', 'age', 'sex', 'unit', 'geo'],  # Mantener estas columnas
    value_vars=[str(year) for year in range(2021, 2024)],  # Años como columnas
    var_name='year',  # Nueva columna para los años
    value_name='rate of adult employment migration (%)'  # Nueva columna para los valores
)

# Mostrar el resultado
df_emplrate.head()
```
